```python
cr['plot_group'] = np.where(cr['record_id'] < 40000, 'Young Adults', 'Older Adults')

current_palette = sns.set_palette("colorblind")

sns.violinplot(x = cr['plot_group'], y = cr['IS'], hue = cr['plot_group'])
plt.title('Interdailty Stability')

from scipy import stats
stats.ttest_ind(ya_df['IS'], oa_df['IS'], equal_var=False)
```

```python
sns.violinplot(x = cr['plot_group'], y = cr['IV'], hue = cr['plot_group'])
plt.title('Intradaily Variability')

from scipy import stats
stats.ttest_ind(ya_df['IV'], oa_df['IV'], equal_var=False)
```

```python
sns.violinplot(x = cr['plot_group'], y = cr['RA'], hue = cr['plot_group'])
plt.title('Relative Amplitude')

from scipy import stats
stats.ttest_ind(ya_df['RA'], oa_df['RA'], equal_var=False)
```

```python
import scipy.io
scipy.io.savemat(os.path.join(work_dir, 'Analysis/rest/rest_wb-cr_cor.mat'), mdict = {'wb_cr_cormat': fc})
```

```python
labels_df = pd.DataFrame(labels, columns = ['labels'], dtype=str)
dmn_mask = labels_df.index[labels_df['labels'].str.contains('Default')]

print('DMN')
print(fc.shape)
fc_dmn = []
fc_dmn = fc[dmn_mask][:,dmn_mask]
print(fc_dmn.shape)

plt.imshow(fc_dmn[:,:,1], interpolation='None', cmap='RdYlBu_r')
scipy.io.savemat(os.path.join(work_dir, 'Analysis/rest/rest_dmn-cr_cor.mat'), mdict = {'dmn_cr_cormat': fc_dmn})
```

```python
fpn_mask = labels_df.index[labels_df['labels'].str.contains('Cont')]

print('FPN')
print(fc.shape)
fc_fpn = []
fc_fpn = fc[fpn_mask][:,fpn_mask]
print(fc_fpn.shape)

plt.imshow(fc_fpn[:,:,1], interpolation='None', cmap='RdYlBu_r')
scipy.io.savemat(os.path.join(work_dir, 'Analysis/rest/rest_fpn-cr_cor.mat'), mdict = {'fpn_cr_cormat': fc_fpn})
```
